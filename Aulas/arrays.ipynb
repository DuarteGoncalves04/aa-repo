```python

D = dt.load_iris()
X = D.data
y = D.target

ax = plt.figure().add_subplot(projection='3d')
ax.plot(X[y==0,0],X[y==0,1],X[y==0,2],'.') #REPRESENTA CLASSE 0
ax.plot(X[y==1,0],X[y==1,1],X[y==1,2],'.') #REPRESENTA CLASSE 1
ax.plot(X[y==2,0],X[y==2,1],X[y==2,2],'.') #REPRESENTA CLASSE 2
ax.set_xlabel('$x$',fontsize=16)
ax.set_ylabel('$y$',fontsize=16)
ax.set_zlabel('$z$',fontsize=16)
plt.show()
```
