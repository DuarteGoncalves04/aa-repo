```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
```

```python
#Discriminantes Lineares  (Slide 5)
x0 = rnd.randn(1000)*np.sqrt(.5) - 1
x1 = rnd.randn(1000)*np.sqrt(.5) + 1
x = np.hstack((x0, x1))
x.shape
```

```python
X = np.vstack((np.ones(2000), x))
y = np.hstack((np.zeros(1000), np.ones(1000)))
plt.plot(y), plt.show()
plt.plot(x0, -np.ones(1000), '.', x1, np.ones(1000), '.'), plt.show()
```

```python
import scipy.linalg as la
Rx = np.dot(X,X.T)
rxy = np.dot(X,y)
w = np.dot(la.inv(Rx), rxy)
plt.plot(x0, -np.ones(1000), '.', x1, np.ones(1000), '.', x, np.dot(w.T, X)), plt.show()
```

```python
t = np.array ([[1,-3],[1,3]])
t = t.T
ty = np.dot(w,t)
tt = np.array([-3,3])
plt.plot(x0, -np.ones(1000), '.', x1, np.ones(1000), '.', tt, ty), plt.show()
```
