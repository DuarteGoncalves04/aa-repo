### Classificador baseados em distâncias

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.spatial.distance as spd
import pickle
```

```python
fName='LAB2distancias001.p'
D = pickle.load(open(fName,'rb'))
```

```python
D.keys()
X = D['X']
y = D['idxClass']
print(X.shape)
print(y.shape)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.axis('scaled')
plt.grid(True)
```

```python
D2 = spd.squareform(spd.pdist(X.T,'euclidean'))
D1 = spd.squareform(spd.pdist(X.T,'cityblock'))
Dc = spd.squareform(spd.pdist(X.T,'cosine'))
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(D2,interpolation='none')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(D1,interpolation='none')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(Dc,interpolation='none')
plt.axis('off')
```

```python
nx = np.sqrt(np.sum(X**2,axis=0))
Xn = X/nx
Dc2 = 1 - np.dot(Xn.T,Xn)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(Dc)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(Dc2)
plt.axis('off')
```

### Classificadores de Distância ao Centroide

```python
fName='LAB2distancias002.p'
D = pickle.load(open(fName,'rb'))
```

```python
D.keys()
X = D['X']
y = D['idxClass']
print(X.shape)
print(y.shape)
```

```python
plt.plot(y)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
```

```python
m1 = np.mean(X[:,y==1],axis=1)
m2 = np.mean(X[:,y==2],axis=1)
m3 = np.mean(X[:,y==3],axis=1)
```

```python
D1 = np.sqrt(np.sum((X-m1[:,np.newaxis])**2,axis=0))
D2 = np.sqrt(np.sum((X-m2[:,np.newaxis])**2,axis=0))
D3 = np.sqrt(np.sum((X-m3[:,np.newaxis])**2,axis=0))
```

```python
D = np.vstack((D1,D2,D3))
ye = np.argmin(D,axis=0)+1
```

```python
plt.plot(y)
plt.plot(ye)
```

```python
from sklearn.metrics import confusion_matrix
MC = confusion_matrix(y,ye)
print(MC)
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.plot(X[0,y!=ye],X[1,y!=ye],'ok',alpha=0.2)
plt.axis('scaled')
```

```python
m1 = np.array([-2,2])
m2 = np.array([2,-2])
m3 = np.array([0,0])
S1 = np.eye(2)*1/4
S2 = np.array([[1/4,0],[0,1/4]])
S3 = np.array([[3,0],[0,1/10]])
```

```python
from numpy.linalg import pinv
S1i = pinv(S1)
S2i = pinv(S2)
S3i = pinv(S3)
```

```python
D1 = np.sqrt(np.sum((X-m1[:,np.newaxis])*np.dot(S1i,X-m1[:,np.newaxis]),axis=0))
D2 = np.sqrt(np.sum((X-m2[:,np.newaxis])*np.dot(S2i,X-m2[:,np.newaxis]),axis=0))
D3 = np.sqrt(np.sum((X-m3[:,np.newaxis])*np.dot(S3i,X-m3[:,np.newaxis]),axis=0))
D = np.vstack((D1,D2,D3))
ye = np.argmin(D,axis=0)+1
MC = confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```

```python
plt.plot(X[0,y==1],X[1,y==1],'.')
plt.plot(X[0,y==2],X[1,y==2],'.')
plt.plot(X[0,y==3],X[1,y==3],'.')
plt.plot(X[0,y!=ye],X[1,y!=ye],'ok',alpha=0.2)
plt.axis('scaled')
```

```python
plt.plot(y)
plt.plot(ye)
```

```python
MC = confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```
