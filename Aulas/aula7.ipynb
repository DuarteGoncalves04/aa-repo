```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dt
import scipy.spatial.distance as spd
```

```python
D = dt.load_iris()
X = D.data
y = D.target
X.shape
```

```python
D1 = spd.squareform(spd.pdist(X,'euclidean'))
D1.shape
plt.imshow(D1, cmap='gray')
```

```python
D = dt.load_digits()
X = D.data
y = D.target
X.shape
```

```python
plt.imshow(X[0,:].reshape(8,8),cmap="gray")
plt.imshow(X[10,:].reshape(8,8),cmap="gray")
plt.imshow(X[100,:].reshape(8,8),cmap="gray")
```

```python
idx = np.argsort(y) #Ordernar tanto os dados como os respetivos labels
y = y[idx]
X = X[idx,:]
D1 = spd.squareform(spd.pdist(X,'euclidean'))
plt.imshow(D1,cmap='gray')
```

```python
D1 = spd.squareform(spd.pdist(X,'cityblock'))
plt.imshow(D1,cmap='gray')
```

```python
D1 = spd.squareform(spd.pdist(X,'cosine'))
plt.imshow(D1,cmap='gray')
```
